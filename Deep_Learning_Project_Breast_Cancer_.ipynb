```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x=df.drop('label',axis=1) #future matrix excpet target variable
y=df['label'] #dependet variable
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15,random_state=2) #split the train test dataset
```

```python
scaler=StandardScaler() #scale the data
x_train_std=scaler.fit_transform(x_train) # fit the scaler in train data
x_test_std=scaler.transform(x_test) #transform the test data
```

**importing tesorflow and keras**

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

**setting up the layers of neural network**

```python
model=keras.Sequential([keras.layers.Flatten(input_shape=(30,)),
                        keras.layers.Dense(20,activation='relu'),
                        keras.layers.Dense(2,activation='sigmoid')]) # Define a Sequential neural network with input flattening, one hidden ReLU layer, and a sigmoid-activated output layer for binary/multiclass classification
```

**compiling neural network**

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

**Training the neural Network**

```python
 history=model.fit(x_train_std,y_train,validation_split=0.1,epochs=100)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training_data','validation_data'],loc='lower right')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training_data','validation_data'],loc='upper right')
plt.show()
```

```python
loss,accuracy=model.evaluate(x_test_std,y_test)
print(accuracy)
```
